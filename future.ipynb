```python
from scipy import interpolate, integrate
import numpy as np
import time


def f1(x, y):
    coef = 1
    return coef * (y**2 + x**2)


discr = 30

X = np.linspace(0, 100, discr)
Y = np.linspace(0, 100, discr)

xg, yg = np.meshgrid(X, Y, indexing="ij")
data = f1(xg, yg)

# вывод данных
for x in X:
    for y in Y:
        print(f"{f1(x, y):8.1f}", end=" ")
    print()

# интерполируемая функция
interp_func = interpolate.RegularGridInterpolator((X, Y), data)

# интегрирование
start = time.monotonic()
print(f"Результат интегрирования обычной функции: {integrate.dblquad(f1, 10, 20, 50, 60)[0]:.2f}. Потрачено {time.monotonic() - start:.2f} секунд")

start = time.monotonic()
print(f"Результат интегрирования интерполированных данных: {integrate.dblquad(lambda x, y: interp_func((x, y)), 10, 20, 50, 60)[0]:.2f}. Потрачено {time.monotonic() - start:.2f} секунд")
```

```python
import numpy as np
from numpy import nan
import pandas as pd
from scipy.interpolate import RegularGridInterpolator
from functools import wraps
```

```python
def dimention_curse():
    """Проклятие размерности"""


def required_kwargs(arguments: tuple[str] | list[str]):
    """Функция с обязательными kwargs параметрами"""

    assert isinstance(arguments, (tuple, list))
    for argument in arguments:
        assert isinstance(argument, str)

    def decorator(function):
        @wraps(function)
        def wrapper(*args, **kwargs):
            parameters = list()
            for argument in arguments:
                if argument not in kwargs:
                    raise ValueError(f"Missing argument: {argument}")
                parameters.append(kwargs[argument])
            result = function(parameters)
            return result

        return wrapper

    return decorator


def create_interpolating_function(df: pd.DataFrame, target: str, method: str = "linear", bounds_error: bool = False, fill_value=nan):
    assert isinstance(df, pd.DataFrame)
    assert target in df.columns

    x, y = df.drop(target, axis=1), df[target]

    axis = tuple(np.unique(ax) for ax in x.values.T)

    y = np.array(y).reshape(*(len(ax) for ax in axis))

    rgi = RegularGridInterpolator(axis, y, method=method, bounds_error=bounds_error, fill_value=fill_value)

    function = required_kwargs(tuple(x.columns))(rgi)

    return function
```
